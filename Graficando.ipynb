```python
import matplotlib.pyplot as plt
import numpy as np
mediasRecompAgentes = [500.0,500.0,500.0,426.6,479.65,482.5,455.45,500.0,468.1,500.0]
qntEpisodiosAgente = [145.0,497.0,96.0,139.0,174.0,226.0,142.0,380.0,102,119]
x1 =  np.arange(len(mediasRecompAgentes))
x2 = [x + 0.25 for x in x1]
plt.bar(x1, mediasRecompAgentes, width=0.25, label = 'Média de Recompensas', color = 'red')
plt.bar(x2, qntEpisodiosAgente, width=0.25, label = 'Episódios levados', color = 'yellow')
agentes = ['Agt1','Agt2','Agt3','Agt4','Agt5', 'Agt6','Agt7', 'Agt8', 'Agt9', 'Agt10']
plt.xticks([x + 0.25 for x in range(len(mediasRecompAgentes))], agentes)
plt.legend()
plt.xlabel("Agentes")
plt.title("Média de Recompensas e episódios levados para conclusão")
plt.show()
```
